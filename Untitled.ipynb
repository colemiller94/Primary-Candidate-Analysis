## Page rank cypher:
CALL algo.pageRank.stream(
  'MATCH (u:User) WHERE exists( (u)-[:TWEETS]-() ) RETURN id(u) as id',
  'MATCH (u1:User)-[:TWEETS]-(:Tweet)-[:RETWEETS]-(:Tweet)-[:TWEETS]-(u2:User) RETURN id(u1) as source, id(u2) as target',
  {graph:'cypher'}
) YIELD nodeId,score with algo.asNode(nodeId) as node, score order by score desc limit 10
RETURN node {.name}, score

## Betweenness cypher:
CALL algo.betweenness.stream()
YIELD nodeId, centrality

MATCH (user:User) WHERE id(user) = nodeId

RETURN user.name AS user,centrality
ORDER BY centrality DESC LIMIT 15;

## Label Propagation cypher:
CALL algo.labelPropagation.stream(null, null,
  {direction: "OUTGOING", iterations: 10})
YIELD nodeId, label
MATCH (user:User) WHERE id(user) = nodeId
RETURN count(user.name) AS size, label ORDER BY label

## Jaccard similarity:
MATCH (u:Candidate)-[*1..2]-(p:User)
WITH u, collect(id(p)) as people
WITH {item:id(u), categories: people} as userData
WITH collect(userData) as data
CALL algo.similarity.jaccard.stream(data)
YIELD item1, item2, count1, count2, intersection, similarity
RETURN algo.asNode(item1).name AS from, algo.asNode(item2).name AS to, intersection, similarity
ORDER BY similarity DESC

```python
def strip_tweets(tweet):
    '''Process tweet text to remove retweets, mentions,links and hashtags.'''
    retweet = r'RT:? ?@\w+:?'
    tweet= re.sub(retweet,'',tweet)
    mention = r'@\w+'
    tweet= re.sub(mention,'',tweet)
    links = r'^(http:\/\/www\.|https:\/\/www\.|http:\/\/|https:\/\/)?[a-z0-9]+([\-\.]{1}[a-z0-9]+)*\.[a-z]{2,5}(:[0-9]{1,5})?(\/.*)?$'
    tweet= re.sub(links,'',tweet)
    tweet_links = r'https:\/\/t\.co\/\w+|http:\/\/t\.co\/\w+'
    tweet=re.sub(tweet_links,'',tweet)	
    tweet_link = r'http\S+'
    tweet=re.sub(tweet_link,'',tweet)
    hashtag = r'#\w+'
    tweet= re.sub(hashtag,'',tweet)
    return tweet
```

```python
import re
query = """\
MATCH p= (c:Candidate)<-[]-(t:Tweet)
where not t :Retweet
with c,collect(t.text) as corpus
return c.screen_name,corpus
"""
df=graph.run(query).to_data_frame()
def strip_tweets(tweet):
    '''Process tweet text to remove retweets, mentions,links and hashtags.'''
    retweet = r'RT:? ?@\w+:?'
    tweet= re.sub(retweet,'',tweet)
    mention = r'@\w+'
    tweet= re.sub(mention,'',tweet)
    links = r'^(http:\/\/www\.|https:\/\/www\.|http:\/\/|https:\/\/)?[a-z0-9]+([\-\.]{1}[a-z0-9]+)*\.[a-z]{2,5}(:[0-9]{1,5})?(\/.*)?$'
    tweet= re.sub(links,'',tweet)
    tweet_links = r'https:\/\/t\.co\/\w+|http:\/\/t\.co\/\w+'
    tweet=re.sub(tweet_links,'',tweet)  
    tweet_link = r'http\S+'
    tweet=re.sub(tweet_link,'',tweet)
    hashtag = r'#\w+'
    tweet= re.sub(hashtag,'',tweet)
    return tweet

def get_toks(corpus):    
    extra_stops = {'amp','be'}
    tokenized_docs = []
    for summary in corpus:
        doc = nlp(strip_tweets(summary.lower()))
        text = []
        for w in doc:
            if not w.is_stop and not w.is_punct and not w.like_num and w.lemma_:
            # we add the lematized version of the word
                if w.lemma_.isalpha() and w.lemma_ not in extra_stops:
                    text.append(w.lemma_)
        tokenized_docs.append(text)
    return [x for x in tokenized_docs if x]

def topic_model(candidatesn,corps):
    bigram = Phrases(corps[candidatesn])
    texts = [bigram[line] for line in corps[candidatesn]]
    
    dictionary = Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]
    return corpus,dictionary
```
